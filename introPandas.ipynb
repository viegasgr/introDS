###AULA 2: Análise exploratória de dados

####Card 1: Valores médios de imóvel e percentual de cada tipo de imóvel presente na base de dados

*Em primeiro lugar, tirar a média de preço dos imóveis*

```python
aluguel.head()
```

```python
aluguel['Valor'].mean() #média de tudo
```

```python
aluguel.groupby('Tipo').mean(numeric_only=True) #esse parâmetro no método mean vai fazer com que só seja tirada a média das colunas numéricas
```

```python
df_preco_tipo = aluguel.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
#os colchetes duplos fazem com q seja gerado um DATAFRAME ao inves de uma série
#o sort_values() vai trazer do menor valor para o maior
```

```python
df_preco_tipo.plot(kind='barh', figsize=(14,10), color='purple') #achamos o valor médio por tipo de imóvel, hurray!
```

```python
#vamos agora explorar por tipo, separando os comerciais dos residenciais
aluguel['Tipo'].unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
aluguel.query('@imoveis_comerciais in Tipo')
#estamos selecionando os valores da coluna Tipo que estão na lista "imoveis_comerciais". o @ é para passar uma variável no query.
```

```python
df_residenciais = aluguel.query('@imoveis_comerciais not in Tipo') #o contrário da seleção anterior
df_residenciais.head()
```

```python
df_residenciais['Tipo'].unique()
```

```python
df_preco_tipo_res = df_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo_res.plot(kind='barh', figsize=(14,10), color='purple') #plota o gráfico das médias de valor dos imóveis residenciais
```

*Agora veremos o percentual de cada imóvel residencial:*

```python
df_residenciais['Tipo'].value_counts(normalize=True) #esse parâmetro faz os números serem mostrados em percentual
df_perc_tipo_res = df_residenciais['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo') #transforma em df e do menor p maior
df_perc_tipo_res.rename(columns={'Tipo': 'Percentuais'}, inplace=True) #muda o nome da coluna para o nome desejado
df_perc_tipo_res
```

```python
df_perc_tipo_res.plot(kind='bar', figsize=(14, 10), color ='green', edgecolor='black', xlabel = 'Tipos', ylabel = 'Percentual'); #visu c grafico
```

A partir da análise do gráfico anterior, é possível perceber que os apts representam mais de 80% dos imóveis residenciais. Sabendo-se que os preços entre casas e apts podem variar muito por suas características distintas, faz sentido querer manter na base de dados apenas os apts.

###AULA 3: Tratando e Filtrando Dados

```python
df_apt.isnull() #informa quais células estão vazias no df
```

```python
df_apt.isnull().sum() #informa QUANTOS dados são nulos por coluna. true=1
```

```python
df_apt = df_apt.fillna(0) #vai substituir todas as células nulas por '0'. funciona bem pq todas as colunas que achou-se nesse df são numéricas.
df_apt
```

```python
'''
Outras maneiras de tratar dados nulos incluem:
> dropna() : apaga a linha que tiver o dado nulo
> fillna(method = 'ffill') : preenche com o valor anterior
> fillna(method = 'bfill') : preenche com o valor posterior
> interpolate() : preencher dados nulos com valores calculados a partir dos vizinhos
'''
```

```python
registros_a_remover = df_apt.query('Valor==0 | Condominio==0').index #pega o index das linhas cujo valor/condominio = 0
```

```python
df_apt.drop(registros_a_remover, axis=0, inplace = True)
#o método axis=0 indica que vai limpar as linhas com os index informados. se fosse axis=1, removeria as colunas
#inplace = True muda direto no df_apt, sem precisar atribuir a df_apt
```

```python
df_apt.drop('Tipo', axis=1, inplace = True) #vai remover a coluna 'Tipo', que não faz sentido uma vez que esse df é só de apts
```

```python
#Para além do .query(), existem formas mais manuais de filtrar dados
selecao1 = df_apt['Quartos'] == 1
df_apt[selecao1] #retorna apenas as linhas em que a selecao1 é True
```

```python
selecao2 = df_apt['Valor'] < 1200.0
df_apt[selecao2] #retorna apenas as linhas em que a selecao2 é True
```

```python
selecao_final = (selecao1) & (selecao2)
df_apt_filtro1 = df_apt[selecao_final] #filtro1: Apts com 1 qt e aluguel menor q 1200
```

```python
selecao3 = (df_apt['Quartos'] >= 2) & (df_apt['Valor'] < 3000) & (df_apt['Area'] > 70)
df_apt_filtro2 = df_apt[selecao3] #filtro2: Apts com no mín 2 qts, aluguel menos q 3000 e area maior q 70m²
```

```python
df_apt.to_csv('dados_apts.csv') #salva o dataframe em um csv com o nome especificado
```

###AULA 4: Manipulando os dados

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

Criando colunas numéricas:

```python
dados['valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['valor_por_ano'] = (dados['valor_por_mes'] * 12) + dados['IPTU']
dados.head()
```

Criando colunas categóricas:

```python
#coluna que descreve o apt
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
#a barra invertida serve para continuar um comando na linha seguinte
dados.head()
```

```python
#coluna que diz se tem suite ou nao
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x>0 else 'Não')
#.apply serve p aplicar uma função em um conjunto de dados.
#lambda são funções anônimas do python, escritas em uma linha apenas e q n vão ser reutilizadas no código. sintaxe básica: lambda {argumentos}: {expressão}

dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
